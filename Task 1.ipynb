```python
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\Cognifyz Internship\Dataset .csv")
df
```

```python
positive_keywords = ['Good', 'Very Good', 'Excellent', 'Average']
negative_keywords = ['poor']
```

```python
positive_count
```

```python
negative_count
```

```python
for review in df['Rating text']:
    review = review.lower()
    for keyword in positive_keywords:
        if keyword in review:
            positive_count += 1
    for keyword in negative_keywords:
        if keyword in review:
            negative_count += 1
```

```python
df['Review Length'] = df['Rating text'].apply(lambda x: len(re.findall(r'\w+', str(x))))
average_review_length = df['Review Length'].mean()
```

```python
plt.scatter(df['Rating text'], df['Review Length'])
plt.xlabel('Rating')
plt.ylabel('Review Length')
plt.title('Relationship between Review Length and Rating')
plt.show()
```
